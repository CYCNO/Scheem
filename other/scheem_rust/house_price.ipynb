```python
from main import Scheem
from sklearn.model_selection import train_test_split # for splitting the data
import pandas as pd
```

```python
df = pd.read_csv("dataset/Housing.csv")
X = df[["area","bedrooms", "bathrooms", "stories"]].values.tolist()[:100]
Y = df[["price"]].values.tolist()[:100]
```

```python
# Normalization of Data
max_features = [max(row[col] for row in X) for col in range(len(X[0]))] 
X_norm = [[row[col] / max_features[col] for col in range(len(row))] for row in X]

max_y = max(val[0] for val in Y) 
Y_norm = [[val[0] / max_y] for val in Y]

# Splitting Data for Training And Testing
X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y_norm, test_size=0.2)
```

```python
model = Scheem([4,1], X_train, Y_train)
model.fit(iter=10000, lr=0.01, log_after=1000)
```

```python
model.accuracy(X_test, Y_test) # Checking Accuracy
```

```python
raw_input = [1000, 3, 2, 1]

# Standarization of Data
norm_input = [raw_input[col] / max_features[col] for col in range(len(raw_input))] 
pred_norm = model.predict(norm_input)
actual_pred = pred_norm[0][0].data * max_y

actual_pred if actual_pred > 0 else 0 # make sure price don't go negative
```
