```python
from scheem import Scheem
```

```python
X = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

Y = [
    [0],
    [1],
    [1],
    [0]
]

activation = ["sigmoid", "sigmoid"]

model = Scheem([2, 2, 1], activation, X, Y)
model.train(iters=10000, lr=1.0, log_after=1000)
```

```python
model.accuracy(X, Y)
```

```python
model.predict([1, 1])
```
