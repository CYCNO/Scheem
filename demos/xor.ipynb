```python
from scheem import Scheem
```

```python
X = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

Y = [
    [0],
    [1],
    [1],
    [0]
]

model = Scheem([2, 2, 1], X, Y, activation_function="sigmoid")
model.fit(iter=10000, lr=1.0)
```

```python
model.accuracy(X, Y)
```

```python
model.predict([1, 0])[0][0].data
```
