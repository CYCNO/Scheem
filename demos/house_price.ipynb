```python
from scheem import Scheem
from sklearn.model_selection import train_test_split # for splitting the data
```

```python
import random

X = []
Y = []

# Generate synthetic houses data (you can generate more if you wish)
for _ in range(20):
    sqft = random.randint(800, 5000)
    beds = random.randint(1, 6)
    age = random.randint(0, 50)
    
    # The secret underlying rule the model needs to learn:
    # Base $50k + ($150 per sqft) + ($20k per bed) - ($2k per year of age)
    true_price = 50.0 + (sqft * 0.15) + (beds * 20.0) - (age * 2.0)
    
    # Add random market noise (-$30k to +$30k) so it isn't perfectly linear.
    # This forces the model to find the line of best fit rather than memorizing.
    market_noise = random.uniform(-30.0, 30.0)
    final_price = true_price + market_noise
    
    X.append([sqft, beds, age])
    Y.append([final_price])

# Check the sizes before training
print(f"Generated {len(X)} rows of data.")
```

```python
# Normalization of Data
max_features = [max(row[col] for row in X) for col in range(len(X[0]))] 
X_norm = [[row[col] / max_features[col] for col in range(len(row))] for row in X]

max_y = max(val[0] for val in Y) 
Y_norm = [[val[0] / max_y] for val in Y]

# Splitting Data for Training And Testing
X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y_norm, test_size=0.2)

model = Scheem([3, 6, 1], X_train, Y_train)
model.fit(iter=10000, lr=0.001)
```

```python
model.accuracy(X_test, Y_test) # Checking Accuracy
```

```python
raw_input = [1000, 3, 5]

# DeNormalization of Data
norm_input = [raw_input[col] / max_features[col] for col in range(len(raw_input))] 
pred_norm = model.predict(norm_input)
actual_pred = pred_norm.matrix[0][0].data * max_y

actual_pred if actual_pred > 0 else 0 # make sure price don't go negative
```
